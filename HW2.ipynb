### Q1 다이아몬드의 색에 따른 가격 분포를 알고싶습니다. ggplot 을 활용해 voilin 플롯, 박스플롯을 그려서 분포, 평균 등을 살펴보세요.

```r
ggplot(data = diamonds, mapping = aes(x = color, y = price)) +
  geom_violin() + geom_boxplot(width = 0.1, color = "grey", alpha = 0.5) + theme_minimal()+
  theme(axis.line= element_line(colour = "black"),
        panel.grid.minor = element_blank(),
        panel.border = element_blank(),
        text = element_text(size = 30), 
        ) +   coord_flip()
```

### Q2 dplyr 함수들을 사용하여, 다이아몬드의 색 별로 평균 분산을 구해보세요. (Hint: group_by, summarize 사용) 

```r
diamonds %>% group_by(color) %>%
summarize(mean = mean(price, na.rm = TRUE), variance = var(price, na.rm = TRUE)) 
```

### Q3 가격이 5000 이하인 다이아몬드들에 대해서, 다이아몬드 색에 따른 가격의 평균과 분산 정보를 Bar plot으로 그려보세요 (bar plot 의 x축은 색, y축은 색에따른 평균가격). 해당 bar plot에 표준편차를 errorbar 로 나타내세요. (Hint: ggplot에서 geom_errorbar 사용)



```r
data = diamonds %>% group_by(color) %>% 
  filter(price <= 5000) %>%
  summarize(mean = mean(price, na.rm = TRUE), variance = var(price, na.rm = TRUE))
```

```r
data
```

```r
ggplot(data = data, mapping = aes(x = color, y = mean)) + 
  geom_bar(stat = "identity") +
  geom_errorbar(aes(x = color, ymin = mean - sqrt(variance), ymax = mean + sqrt(variance)), colour="orange")
```

```r
ggplot(data = data, mapping = aes(x = color, y = variance)) + 
  geom_bar(stat = "identity") +
  geom_errorbar(aes(x = color, ymin = variance - sqrt(variance), ymax = variance + sqrt(variance)), colour="orange")
```
